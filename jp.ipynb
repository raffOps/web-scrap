```python
import requests as re
from bs4 import BeautifulSoup as bs
import pandas as pd
import time
```

```python
url = "https://busca.movidaseminovos.com.br/filtros/class/usado?sort=11&page={}"
r = re.get(url)
soup = bs(r.text, "lxml")
```

```python
nomes_coluna = "empresa modelo preco kilometragem ano".split()
db = pd.DataFrame(columns=nomes_coluna)
```

```python
indice_pagina = 1
while True:
    r = re.get(url.format(indice_pagina))
    soup = bs(r.text, "lxml")
    if soup.find_all("div",{"class":"nm-not-found-message"}):
        break
    else:
        for carro in soup.find_all("div", {"class": "nm-product-info"}):
            dados.append("Movidas")
            dados.append(carro.find_all("a")[0]["title"])
            dados.append(carro.find("div").text.split("\n")[1].split()[1])
            dados.append(carro.find_all("span")[0].text)
            dados.append(carro.find_all("span")[2].text)
            adicao = {nomes_coluna[index]:dados[index] for index in range(5)}
            db = db.append(adicao, ignore_index=True)
            del(dados[:])
        print(db.shape[0])
        indice_pagina = indice_pagina + 1
        time.sleep(2)
```
